**Random Forest**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")

from google.colab import drive
drive.mount('/content/drive')
```

```python
df = pd.read_csv("/content/drive/MyDrive/data/random+forest+regression+dataset.csv", sep=";")
```

```python
df
```

```python
x = df.iloc[:,0].values.reshape(-1,1) 
y = df.iloc[:,1].values.reshape(-1,1)#values methodu ile pandas series array'e çavrilip, reshape methodu ile 1 sütundan oluştuğu sklearn'e bildirilir.
```

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(n_estimators=100 , random_state=42) #RandomForestRegressor modeli rf adlı variable'a eşitlenir.
```

```python
rf.fit(x,y) #x ve y değerleri kullanılarak model fit edilir.
```

```python
y_head = rf.predict(x)  #x değerlerine karşılık gelen y değerleri predict edilir.
```

Burada prediction yapılırken eğitilen data kullanılmaktadır. Aslında bir test verisinin olması lazımdır.

```python
y_head
```

Bu değerler prediction sonucu bulunan değerlerdir.

Bu değerlerin gerçek değerlere göre doğru olup olmadığını görmek için r^2 kullanılır.

```python
from sklearn.metrics import r2_score
print("r^2 score: ",r2_score(y,y_head))
```

r^2 score 0.98 çıkmıştır. 1'e çok yakın olduğu için yapılan prediction'ın çoğunun doğru olduğu görülür. Model ile data uyumludur.

**Linear Regression**

```python
df = pd.read_csv("/content/drive/MyDrive/data/linear_regression_dataset.csv", sep=";")
```

```python
df.head()
```

```python
x = df.deneyim.values.reshape(-1,1)
y = df.maas.values.reshape(-1,1) #values methodu ile pandas series array'e çavrilip, reshape methodu ile 1 sütundan oluştuğu sklearn'e bildirilir.
```

```python
from sklearn.linear_model import LinearRegression
linear_reg = LinearRegression() #LinearRegression modeli linear_reg adlı variable'a eşitlenir.
linear_reg.fit(x,y)             #line fit edilir.
```

```python
y_head = linear_reg.predict(x) #tahmin edilen maaş değerleridir.
plt.plot(x, y_head, color="red")
plt.scatter(x,y)
plt.show()
```

Plot incelendiğinde modelin predict ettiği line'ın verilerin çoğunun üzerinden geçtiği görülür. Modelin doğru bir tahmin yaptığı gözle bakınca da anlaşılabilir.

```python
from sklearn.metrics import r2_score
print("r^2 score: ",r2_score(y,y_head))
```

0.97 r^2 score modelin yaptığı predictionların doğru olduğunu gösterir.
