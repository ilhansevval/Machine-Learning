# **Multiple Linear Regression**

Birden çok bağımsız değişkenler (x) ile bir tane bağımlı değişken (y) arasındaki ilişkiyi modeller.

Linear Regressionda bir tane bağımsız değişken bulunurken multiple linear regression'da birden çok bağımsız değişken bulunur.

1_agw6iTzCguj_qvtIP-mRIg.png

Multiple Linear Regression formülündeki y, y eksenidir. Datadaki maaş sütunudur.

x değişkenleri ise deneyim ve yaş sütunlarıdır.

x değişkenleri(bağımsız değişkenler) kullanarak y değişkeni(bağımsız değişken) tahmin edilmeye çalışılacaktır.

y = b0+ b1.x1 +b2.x2 -> multiple linear regression denklemi dataya uygulandığında;

maas= b0+ b1.deneyim + b2.yas formülü elde edilir.

Buna göre deneyim ve yas(bağımsız değişkenler) üzerinden maas(bağımlı değişken)

tahmin edilmeye çalışır.

Amaç Linear Regression'da olduğu gibi MSE değerini düşük tutmaktır. Çünkü MSE değeri ne kadar düşük çıkarsa modeldeki hata oranı o kadar azdır ve doğru bir tahmin gerçekleştirildiği anlamına gelir.

MSE'yi düşük tutabilmek için y = b0 + b1.x+b2.x+bn.xn formülündeki en uygun b0 ve b1,b2...bn değerleri bulunmalıdır. Çünkü bu formülden çıkan y değeri residual hesaplanırken kullanırken y değeridir.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression

from google.colab import drive
drive.mount('/content/drive')
```

```python
df = pd.read_csv("/content/drive/MyDrive/data/multiple_linear_regression_dataset.csv", sep=";")
```

```python
df.head()
```

```python
x = df.iloc[:,[0,2]].values       #deneyim ve maas sütunundaki değerler alınır. 
y = df.maas.values.reshape(-1,1)  #maas sütunundaki değerler alınarak 1 sütundan oluştuğu belirtilir.
```

```python
multiple_linear_regression = LinearRegression()  #LinearRegression modeli multiple_linear_regression adlı variable'a eşitlenir.
multiple_linear_regression.fit(x,y)              #x ve y değerleri kullanılarak line fit edilir.

print("b0: ", multiple_linear_regression.intercept_)  #multiple linear regressionda b0 bağımlı değişken intercept methodu ile çağırılır.
print("b1,b2: ",multiple_linear_regression.coef_)     #b1 ve b2 bağımsız değişken coef ile çağırılır.
```

y = b0+ b1.x1 +b2.x2 regression denklemiyle datayı fit ettikten sonra;

maas = 10376+ 1525.deneyim -416.yas formülü elde edilir.

```python
multiple_linear_regression.predict(np.array([[10,35],[5,35]]))
#oluşturulan modele göre 10 yıllık deneyimi olan 35 yaşında birinin maaşı ve 5 yıllık deneyimi olan 35 yaşında birinin maaşı tahmin edilmiştir.
```
